```python
def get_full_name(first, name):
    return first + ' ' + name
r = get_full_name('佐藤', '太郎')
print(r)
```

```python
r = get_full_name(name = '太郎', first = '佐藤')
print(r)
```

```python
def half_value(x):
    return x / 2
r = half_value(2)
print(r)
```

```python
half_value_v2 = lambda x: x / 2

r2 = half_value_v2(2)
print(r2)
```

```python
def f(x):
    if x % 2 == 0:
        return x
    else:
        return '2で割り切れません！！'
    
r = f(2)
print(r)
```

```python
f2 = lambda x: x if x % 2 == 0 else '2で割り切れません！！'
r2 = f2(3)
print(r2)
```

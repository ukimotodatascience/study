# 7 品詞毎の分類
先ほどの文章に形態素解析を行い、その要素ごとに以下の項目をもつ辞書を作成し、全要素分をもつリストresultsを作成してください

### 辞書型オブジェクト
作成する辞書型オブジェクトは次の4つの項目のキーと値をもちます
- 表層形（一番最初）：「はじめまして」,「、」,「こちら」
- 品詞（\tの右側）：「感動詞」,「記号」,「名詞」
- 品詞1：「*」,「*」,「代名詞」
- 基本形：「はじめまして」,「、」,「こちら」

```python
tagger = MeCab.Tagger()
parsed_txt = tagger.parse(txt)
elements = parsed_txt.split('\n')[:-2]
elements
```

```python
results = []

for element in elements:
    parts = element.split(',')
    surface_pos, pos1, base = parts[0], parts[1], parts[-3]
    surface, pos = surface_pos.split('\t')
    results.append(dict(表層形=surface, 基本形=base, 品詞=pos, 品詞1=pos1))

results
```

# 8 動詞の抽出
抽出した品詞群の中から重複のないように動詞の表層形のみ（surface_verbs）および動詞の基本形（base_verbs）を抽出してください

```python
result = results[0]

surface_verbs = set()
base_verbs = set()
for result in results:
    if result['品詞'] == '動詞':
        surface_verbs.add(result['表層形'])
        base_verbs.add(result['基本形'])

print('表層形:', surface_verbs)
print('基本形:', base_verbs)
```

# 9 単語の出現頻度の算出
辞書のキーに各単語、バリューには、対応する単語の出現頻度が入るような辞書を作成してください

```python
from collections import defaultdict
```

```python
word_freq = defaultdict(int)

for result in results:
    if result['品詞'] != '記号':
        word_freq[result['基本形']] += 1

word_freq
```

# 12 N-gram（文字単位）
例文をbi-gramとtri-gramにして表示してください

例文）今日は晴れです。

```python
txt = '今日は晴れです。'

# bi-gram
for i in range(len(txt)-1):
    print(txt[i:i+2])
```

```python
# tri-gram
for i in range(len(txt)-2):
    print(txt[i:i+3])
```

# 13 N-gram（単語単位）
こちらの例文を一度単語区切りにしたのちbi-gramとtri-gramにして表示してください

例文）Pythonを一緒に学びましょう。

```python
txt = 'Pythonを一緒に学びましょう。'

tagger = MeCab.Tagger('-Owakati')
words = tagger.parse(txt).split(' ')[:-1]

# bi-gram
for i in range(len(words)-1):
    print(''.join(words[i:i+2]))
```

```python
# tri-gram
for i in range(len(words)-2):
    print(''.join(words[i:i+3]))
```

# 14 N-gramの関数化
文字および単語区切りのN-gramを出力する関数ngram()を作成してください

全体をリスト型で返しましょう

こんにちは、プログラミング講師のいまにゅです。

```python
n = 3

[words[i:i+n] for i in range(len(words)-n-1)]
```

```python
txt = '今日は晴れです。'
[txt[i:i+n] for i in range(len(txt)-n-1)]
```

```python
def ngram(text, n):
    return [text[i:i+n] for i in range(len(txt)-n-1)]
```

```python
ngram(words, 3)
```

```python
ngram(txt, 3)
```

# 19 k-meansでクラスタリング
先ほどの単語ベクトルのモデルから1000単語を抽出し、k-meansクラスタリングをクラスタ数k=100として実行してください

また、それぞれのクラスタ内にある単語の先頭10個をそれぞれ出力してください

```python
max_vocab = 1000
vocab = list(model.key_to_index.keys())[:max_vocab]
vectors = [model.get_vector(word) for word in keys]
```

```python
from sklearn.cluster import KMeans

n_clusters = 100
kmeans_model = KMeans(n_clusters=n_clusters, verbose=1, random_state=42)
kmeans_model.fit(vectors)
```

```python
cluster_labels = kmeans_model.labels_
cluster_to_words = defaultdict(list)
for cluster_id, word in zip(cluster_labels, vocab):
    cluster_to_words[cluster_id].append(word)
```

```python
for words in cluster_to_words.values():
  print(words[:10])
```

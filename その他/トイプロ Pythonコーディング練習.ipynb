# 日替わりランチ

トイプロ食堂では平日限定で日替わり定食を提供しています。

今日の曜日（today）が与えられるので、今日の日替わり定食の内容を表示してください！

todayが平日でなければ、何も表示しません。

```python
#自作
today = input()

if today == '月':
    print('ハンバーグ定食')
elif today == '火':
    print('から揚げ定食')
elif today == '水':
    print('生姜焼き定食')
elif today == '木':
    print('焼き魚定食')
else:
    print('コロッケ定食')
```

```python
#サプー
menu_map = {
    '月':'ハンバーグ定食',
    '火':'から揚げ定食',
    '水':'生姜焼き定食',
    '木':'焼き魚定食',
    '金':'コロッケ定食'
}

today = '月'
if today in menu_map.keys():
    print(menu_map[today])
```

# クイズ大会の順位

ある所でクイズ大会が開かれました！

一定時間内に、より多くの問題を解いた人がより上位となります！

もしも、同じ数の問題を解いた人がいれば、より短い時間で解いた人が上位となります！

リスト『entries』には参加した人の『名前』『正解数』『時間』の入った辞書が与えられます！

『名前』が『name』の人は何位だったでしょう？

## 制約
リスト『entries』の『名前』のどれかに『name』が必ず入っている。
同率順位だった場合は、同じ順位の人は無視して、『上位にいる人の数+1』の順位にして表示してください。

```python
#自作
```

```python
#サプー
entries = [
    {'名前':'足利','正解数':100,'時間':50},
    {'名前':'織田','正解数':120,'時間':40},
    {'名前':'明智','正解数':100,'時間':10},
    {'名前':'豊臣','正解数':120,'時間':40},
    {'名前':'徳川','正解数':150,'時間':60}
]

name = '明智'
target_entry = {}
for entry in entries:
    if entry['名前'] == name:
        target_entry = entry

higher_ranker_cnt = 0
for entry in entries:
    if entry['正解数'] > target_entry['正解数']:
        higher_ranker_cnt += 1
    elif (entry['正解数'] == target_entry['正解数'] and entry['時間'] < target_entry['時間']):
        higher_ranker_cnt += 1

print(f'{higher_ranker_cnt + 1}位')
```

```python
#サプー
entries = [
    {'名前':'足利','正解数':100,'時間':50},
    {'名前':'織田','正解数':120,'時間':40},
    {'名前':'明智','正解数':100,'時間':10},
    {'名前':'豊臣','正解数':120,'時間':40},
    {'名前':'徳川','正解数':150,'時間':60}
]

name = '明智'
target_entry = list(filter(
    lambda x: x['名前'] == name, entries))[0]
higher_ranker = list(filter(
    lambda x: x['正解数'] > target_entry['正解数'] or (
        x['正解数'] == target_entry['正解数'] and
        x['時間'] < target_entry['時間']
    ), entries
))

print(f'{len(higher_ranker) + 1}位')
```

# 塩と水

塩水に追加で塩と水を加えてよく混ぜました。

元の塩水の重さがaグラムで、塩の濃度がb%でした。

加えた水の重さxグラムと塩の重さyグラムが与えられるので、塩の濃度は何%になったでしょう？

## 制約
1 ≤ a ≤10⁴

1 ≤ b ≤ 28

0 ≤ x ≤ 10⁴

0 ≤ y ≤ 10⁴

a,b,x,yは整数で与えられる。

単位は%で、小数点以下を切り捨てて表示する。

表示は必ず28%以下になる。

```python
#自作
```

```python
#サプー
a = 100
b = 10
x = 90
y = 10

before_salt_g = a * b / 100
before_water_g = a - before_salt_g

after_salt_g = before_salt_g + y
after_water_g = before_water_g + x

after_concent = after_salt_g / (after_salt_g + after_water_g) * 100
print(int(after_concent))
```

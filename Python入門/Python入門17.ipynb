```python
try:
    add_num = lambda a, b: a + b
    r = add_num(1, '2')
except:
    print('例外が発生しました')

print('例外処理が終わりました。')
```

```python
add_num = lambda a, b: a + b
try:
    r = add_num(1, '2')
    print(r)
except TypeError as e:
    print(e)
    print('例外が発生しました')

print('例外処理が終わりました。')
```
